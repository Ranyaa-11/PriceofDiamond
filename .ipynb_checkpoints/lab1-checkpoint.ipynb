```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(42)
num_data=np.random.randint(20,80,size=95).tolist()
outliers=[150,160,170,7,10]
num_data.extend(outliers)
categories=['a','b','c','d']
cat_data=np.random.choice(categories,size=100)
df=pd.DataFrame({'num_col':num_data,'cat_col':cat_data})
```

```python
q1=df['num_col'].quantile(0.25)
q3=df['num_col'].quantile(0.75)
iqr=q3-q1
lower_bnd=q1-1.5*iqr
upp_bnd=q1+1.5*iqr
outliers=df[(df['num_col']<lower_bnd)|(df['num_col']>upp_bnd)]['num_col']
print("outl\n",outliers)
```

```python
# Step 1: Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# Step 2: Load the dataset (downloaded from Kaggle)
df = pd.read_csv(r"C:\Users\kshem\Downloads\IMDB Dataset.csv")

# Step 3: Inspect the data
print("Dataset shape:", df.shape)

print(df)

# Step 4: Encode labels (Positive → 1, Negative → 0)
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})

# Step 5: Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    df['review'], df['sentiment'], test_size=0.2, random_state=42
)

# Step 6: Convert text to numerical features using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Step 7: Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Step 8: Evaluate the model
y_pred = model.predict(X_test_tfidf)
print(classification_report(y_test, y_pred))
```
